# Preguntas y Respuestas

```python
import random
import unicodedata

# FunciÃ³n para eliminar tildes y normalizar texto por librerÃ­a unicodedata
def normalizar_tildes(palabra):
    # Crear una lista vacÃ­a para almacenar los caracteres
    caracteres_sin_tildes = []
    
    # "Normalizar" la palabra (descomponer los caracteres)
    palabra_normalizada = unicodedata.normalize('NFD', palabra)
    
    # Recorrer cada caracter de la palabra normalizada
    for caracter in palabra_normalizada:
        # Si el caracter no es una marca de acento (no es 'Mn'), lo agregamos a la lista
        if unicodedata.category(caracter) != 'Mn':
            caracteres_sin_tildes.append(caracter)
    
    # Unir los caracteres
    return ''.join(caracteres_sin_tildes).title().strip()

# FunciÃ³n para comparar texto de forma insensible a tildes y mayÃºsculas/minÃºsculas
def comparar_palabras(palabra1, palabra2):
    return normalizar_tildes(palabra1) == normalizar_tildes(palabra2)

# FunciÃ³n para solicitar el nombre del jugador
def solicitar_nombre_jugador():
    print("Â¡BIENVENID@ AL JUEGO DE PREGUNTAS Y RESPUESTAS SOBRE GEOGRAFÃA ğŸŒ!")
    print("Puedes responder un mÃ¡ximo de 5 preguntas CORRECTAS âœ… y un mÃ¡ximo de 3 INCORRECTAS âŒ. Â¡Â¡Suerte!! ğŸ€")
    
    nombre = input("Â¿CÃ³mo te llamas?").strip().title()
    if not nombre:
        nombre = "Jugador AnÃ³nimo"
    print(f"\nÂ¡Â¡Hola {nombre}!! ğŸ‘ğŸ» Â¡Espero que te lo pases bien y te sepas todas las respuestas! ğŸ˜Š")
    return nombre

# DeclaraciÃ³n del diccionario con las preguntas y las respuestas
preguntas_respuestas = {
    "Â¿DÃ³nde se encuentra el Monte Fuji â›°?": "JapÃ³n",
    "Â¿CuÃ¡l es el rÃ­o mÃ¡s grande de EspaÃ±a?": "Ebro",
    "Â¿CuÃ¡l es el paÃ­s mÃ¡s pequeÃ±o del mundo ğŸ—º?": "Ciudad del Vaticano",
    "Â¿En quÃ© paÃ­s se encuentra la Torre Eiffel ğŸ—¼?": "Francia",
    "Â¿QuÃ© Comunidad AutÃ³noma se encuentra en la meseta de la PenÃ­nsula IbÃ©rica?": "Madrid",
    "Â¿CuÃ¡l es la montaÃ±a ğŸŒ‹ mÃ¡s alta de EspaÃ±a?": "Teide",
    "Â¿CuÃ¡ntas Comunidades AutÃ³nomas hay en EspaÃ±a?": "17",
    "Â¿En quÃ© ciudad de Estados Unidos estÃ¡ Walt Disney World ğŸ¢ğŸ¡?": "Orlando",
    "Â¿QuÃ© paÃ­s es conocido por la pizza ğŸ• y la pasta ğŸ?": "Italia",
    "Â¿QuÃ© ciudad es considerada la mÃ¡s antigua del mundo ğŸ—º?": "Damasco"
}

# FunciÃ³n del juego
def juego_preguntas_respuestas():
    nombre_jugador = solicitar_nombre_jugador() # Llamada de la funciÃ³n 
    
    # InicializaciÃ³n de variables
    intentos_incorrectos = 0 
    respuestas_correctas = 0 
    contador_rondas = 1
    max_respuestas_incorrectas = 3
    max_respuestas_correctas = 5
    preguntas_hechas = []

    while intentos_incorrectos < max_respuestas_incorrectas and respuestas_correctas < max_respuestas_correctas:
        pregunta_visible = seleccionar_pregunta(preguntas_hechas)

        # Mostrar el nÃºmero de la ronda actual, la pregunta y la respuesta
        print(f"\n---------- RONDA {contador_rondas} ----------")
        print(f"Pregunta: {pregunta_visible}")
        respuesta_jugador = input("Tu respuesta: ").strip().title()
        print(f"Respuesta: {respuesta_jugador}")
        preguntas_hechas.append(pregunta_visible)

        # Usar la funciÃ³n comparar_texto para verificar la respuesta
        if comparar_palabras(respuesta_jugador, preguntas_respuestas[pregunta_visible]):
            print("\nÂ¡Esa respuesta es CORRECTA! ğŸ˜")
            respuestas_correctas += 1
        else:
            print(f"\nOoh ... respuesta incorrecta ğŸ˜¢. La respuesta correcta era: {preguntas_respuestas[pregunta_visible].upper()}. Â¡La siguiente seguro que la sabes! ğŸ’ªğŸ¼")
            intentos_incorrectos += 1

        print(f"Respuestas correctas âœ…: {respuestas_correctas}")
        print(f"Respuestas incorrectas âŒ: {intentos_incorrectos}")

        # Incrementar el contador de rondas
        contador_rondas += 1

    # ValidaciÃ³n para ver el ganador 
    if respuestas_correctas >= max_respuestas_correctas:
        print(f"\nğŸ‰ Â¡Â¡FELICIDADES, {nombre_jugador}!! ğŸ‰ Sabes un montÃ³n de geografÃ­a ğŸŒ.")
    elif intentos_incorrectos >= max_respuestas_incorrectas:
        print(f"\n{nombre_jugador}, has fallado tres preguntas ğŸ˜”, se te han acabado el nÃºmero de intentos fallidos.")
        print("Seguro que la prÃ³xima vez se te da mejor. Â¡Suerte si lo vuelves a intentar! ğŸ¤—ğŸ˜˜ ")
    
# FunciÃ³n para seleccionar una pregunta aleatoria
def seleccionar_pregunta(preguntas_hechas):
    # Filtra las preguntas que no han sido hechas y selecciona una aleatoria si existen
    preguntas_disponibles = set(preguntas_respuestas) - set(preguntas_hechas)
    return random.choice(list(preguntas_disponibles)) #if preguntas_disponibles else None

# EjecuciÃ³n del juego
juego_preguntas_respuestas()
```
