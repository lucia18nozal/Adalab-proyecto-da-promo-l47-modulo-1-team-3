```python
import random
import unicodedata

class JuegoPreguntasGeografia:
    def __init__(self):
        # DeclaraciÃ³n del diccionario con las preguntas y las respuestas e inicializaciÃ³n de variables
        self.preguntas_respuestas = {
            "Â¿DÃ³nde se encuentra el Monte Fuji â›°?": "JapÃ³n",
            "Â¿CuÃ¡l es el rÃ­o mÃ¡s grande de EspaÃ±a?": "Ebro",
            "Â¿CuÃ¡l es el paÃ­s mÃ¡s pequeÃ±o del mundo ğŸ—º?": "Ciudad del Vaticano",
            "Â¿En quÃ© paÃ­s se encuentra la Torre Eiffel ğŸ—¼?": "Francia",
            "Â¿QuÃ© Comunidad AutÃ³noma se encuentra en la meseta de la PenÃ­nsula IbÃ©rica?": "Madrid",
            "Â¿CuÃ¡l es la montaÃ±a ğŸŒ‹ mÃ¡s alta de EspaÃ±a?": "Teide",
            "Â¿CuÃ¡ntas Comunidades AutÃ³nomas hay en EspaÃ±a?": "17",
            "Â¿En quÃ© ciudad de Estados Unidos estÃ¡ Walt Disney World ğŸ¢ğŸ¡?": "Orlando",
            "Â¿QuÃ© paÃ­s es conocido por la pizza ğŸ• y la pasta ğŸ?": "Italia",
            "Â¿QuÃ© ciudad es considerada la mÃ¡s antigua del mundo ğŸ—º?": "Damasco"
        }
        self.nombre_jugador = ""
        self.preguntas_hechas = []
        self.intentos_incorrectos = 0
        self.respuestas_correctas = 0
        self.contador_rondas = 1
        self.max_respuestas_incorrectas = 3
        self.max_respuestas_correctas = 5

    # MÃ©todo para eliminar tildes y normalizar texto por librerÃ­a unicodedata
    def normalizar_tildes(self, palabra):
            # Crear una lista vacÃ­a para almacenar los caracteres
            caracteres_sin_tildes = []
            # "Normalizar" la palabra (descomponer los caracteres)
            palabra_normalizada = unicodedata.normalize('NFD', palabra)
            # Recorrer cada caracter de la palabra normalizada
            for caracter in palabra_normalizada:
                # Si el caracter no es una marca de acento (no es 'Mn'), lo agregamos a la lista
                if unicodedata.category(caracter) != 'Mn':
                    caracteres_sin_tildes.append(caracter)
            # Unir los caracteres
            return ''.join(caracteres_sin_tildes).title().strip()

    # FunciÃ³n para comparar texto de forma insensible a tildes y mayÃºsculas/minÃºsculas
    def comparar_palabras(self, palabra1, palabra2):
            return self.normalizar_tildes(palabra1) == self.normalizar_tildes(palabra2)

    # FunciÃ³n para solicitar el nombre del jugador
    def solicitar_nombre_jugador(self):
            # Mensaje de inicio
            print("Â¡BIENVENID@ AL JUEGO DE PREGUNTAS Y RESPUESTAS SOBRE GEOGRAFÃA ğŸŒ!")
            print("Puedes responder un mÃ¡ximo de 5 preguntas CORRECTAS âœ… y un mÃ¡ximo de 3 INCORRECTAS âŒ. Â¡Â¡Suerte!! ğŸ€")

            # Nombre del jugador voluntario o automÃ¡tico
            nombre = input("Â¿CÃ³mo te llamas?").strip().title()
            if not nombre:
                nombre = "Jugador AnÃ³nimo"
            self.nombre_jugador = nombre

            print(f"\nÂ¡Â¡Hola {nombre}!! ğŸ‘ğŸ» Â¡Espero que te lo pases bien y te sepas todas las respuestas! ğŸ˜Š")
    
    # FunciÃ³n para seleccionar una pregunta aleatoria que no haya sido hecha ya
    def seleccionar_pregunta(self):
        # Filtra las preguntas que no han sido hechas y selecciona una aleatoria si existen
        preguntas_disponibles = set(self.preguntas_respuestas) - set(self.preguntas_hechas)
        return random.choice(list(preguntas_disponibles))

    # FunciÃ³n para ejecutar el juego
    def jugar(self):
        self.solicitar_nombre_jugador() 

        while self.intentos_incorrectos < self.max_respuestas_incorrectas and self.respuestas_correctas < self.max_respuestas_correctas:
            
            pregunta_visible = self.seleccionar_pregunta()
            # Mostrar el nÃºmero de la ronda actual, la pregunta y la respuesta
            print(f"\n\n---------- RONDA {self.contador_rondas} ----------")
            print(f"Pregunta: {pregunta_visible}")
            respuesta_jugador = input("Tu respuesta: ").strip().title()
            print(f"Respuesta: {respuesta_jugador}")
            self.preguntas_hechas.append(pregunta_visible)

            # Usar la funciÃ³n comparar_palabras para verificar la respuesta
            if self.comparar_palabras(respuesta_jugador, self.preguntas_respuestas[pregunta_visible]):
                print("\nÂ¡Esa respuesta es CORRECTA! ğŸ˜")
                self.respuestas_correctas += 1
            else:
                print(f"\nOoh ... respuesta incorrecta ğŸ˜¢. La respuesta correcta era: {self.preguntas_respuestas[pregunta_visible].upper()}. Â¡La siguiente seguro que la sabes! ğŸ’ªğŸ¼")
                self.intentos_incorrectos += 1

            print(f"\nRespuestas correctas âœ…: {self.respuestas_correctas}")
            print(f"Respuestas incorrectas âŒ: {self.intentos_incorrectos}")
            # Incrementar el contador de rondas
            self.contador_rondas += 1

        # ValidaciÃ³n para ver el ganador 
        if self.respuestas_correctas >= self.max_respuestas_correctas:
            print(f"\n\nğŸ‰ Â¡Â¡FELICIDADES, {self.nombre_jugador}!! ğŸ‰ Sabes un montÃ³n de geografÃ­a ğŸŒ.")
        elif self.intentos_incorrectos >= self.max_respuestas_incorrectas:
            print(f"\n{self.nombre_jugador}, has fallado tres preguntas ğŸ˜”, se te han acabado el nÃºmero de intentos fallidos.")
            print("Seguro que la prÃ³xima vez se te da mejor. Â¡Suerte si lo vuelves a intentar! ğŸ¤—ğŸ˜˜")

# EjecuciÃ³n del juego
if __name__ == "__main__":
    juego = JuegoPreguntasGeografia()
    juego.jugar()
```
